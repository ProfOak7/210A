```python
# importing the required module 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel (r'./Spring2022_SLO_unnamed.xlsx', index_col=0) 
df.describe()
```

```python
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':300})

# Plot Histogram on x
x1 = df.SLO1_percent
plt.hist(x1, bins = 15, edgecolor='k')
plt.gca().set(title='SLO 1 - Frequency Distribution', ylabel='Frequency');
plt.xlim(0,100)

plt.axvline(x1.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x1.mean()))
```

```python
kwargs = dict(hist_kws={'alpha':.4}, kde_kws={'linewidth':3})

plt.figure(figsize=(10,7), dpi= 300)
sns.distplot(x1, bins = 15, color="dodgerblue", label="Original", **kwargs)
plt.axvline(x1.mean(), color='dodgerblue', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x1.mean()), color="dodgerblue")
plt.gca().set(title='SLO 1 - Score Density', ylabel='Density', xlabel='Score');
```

#### SLO 1 Observations: 
The results are not normally distributed as there are two unique populations observed in this assessment. There is a population of students that strongly understand and can implement the scientific method (scoring at or near perfect on this assessment). There is a second population of students who generally understand, but mix up a few terms in their application. There are no major outliers in this assessment.

```python
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':300})

# Plot Histogram on x
x2 = df.SLO3_percent
plt.hist(x2, bins = 25, edgecolor='k')
plt.gca().set(title='SLO 3 - Frequency Distribution', ylabel='Frequency');
plt.xlim(0,100)

plt.axvline(x2.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x2.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x2.mean()))
```

```python
kwargs = dict(hist_kws={'alpha':.4}, kde_kws={'linewidth':3})

plt.figure(figsize=(10,7), dpi= 300)
sns.distplot(x2, bins = 25, color="orange", label="Original", **kwargs)
plt.axvline(x2.mean(), color='orange', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x2.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x2.mean()), color="orange")
plt.gca().set(title='SLO 3 - Score Density', ylabel='Density', xlabel='Score');
```

#### SLO 3 Observations: 
The results are more normally distributed, however a few outliers exist. One of the outliers (near 40) is explainable as this is a studet currently failing the course. However, the second outlier (near 70) is troubling as this is a student who has performed well througout the course but has not have retained these basic concepts. Overall, a majority of students seem to have retained and understad these basic concepts well.

```python
kwargs = dict(hist_kws={'alpha':.4}, kde_kws={'linewidth':3})

plt.figure(figsize=(10,7), dpi= 600)
sns.distplot(x1, bins = 15, color="dodgerblue", label="SLO 1", **kwargs)
plt.axvline(x1.mean(), color='dodgerblue', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*.77, max_ylim*0.75, 'Mean: {:.2f}'.format(x1.mean()), color="dodgerblue")

sns.distplot(x2, bins = 25, color="orange", label="SLO 3", **kwargs)
plt.axvline(x2.mean(), color='orange', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x2.mean()*1.025, max_ylim*0.75, 'Mean: {:.2f}'.format(x2.mean()), color="orange")
plt.xlim(0,120)
plt.legend();
plt.gca().set(title='210A Spring 2022 - Comparison of SLO 1 and SLO 3 Score Density', ylabel='Density', xlabel='Score');
```
